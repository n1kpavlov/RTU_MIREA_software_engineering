–†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω–∏–µ –∏ –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –º–∞—Å—Å–∏–≤–∞, —Å—Ä–∞–≤–Ω–∏—Ç–µ
—Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã, –∫–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –æ –∑–Ω–∞—á–µ–Ω–∏—è—Ö?
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ–Ω–∏—è –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞.

```python
import random
import numpy as np
import matplotlib.pyplot as plt

x = [random.random() for i in range(10)]
avg = np.average(x)
med = np.median(x)

print(x)
print('–°—Ä–µ–¥–Ω–µ–µ: ', avg)
print('–ú–µ–¥–∏–∞–Ω–∞: ', med)

if (avg > med):
  print('—Å—Ä–µ–¥–Ω–µ–µ > –º–µ–¥–∏–∞–Ω—ã')
else:
  print('—Å—Ä–µ–¥–Ω–µ–µ < –º–µ–¥–∏–∞–Ω—ã')

plt.grid()
plt.xlabel("–ù–æ–º–µ—Ä —ç–ª–µ–º–µ–Ω—Ç–∞")
plt.ylabel("–ó–Ω–∞—á–µ–Ω–∏–µ")
plt.title("–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ–∏–≤–∞–∏—è")
plt.scatter(range(10), x)
```

4.3.2 –ó–∞–¥–∞–Ω–∏–µ: –î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è f. –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π
—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π

```python
import numpy as np
import matplotlib.pyplot as plt

y = []
for x in range(1, 11):
  y.append(np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x*x))
  / abs(1 - np.sin(x) ** 3) + np.log(np.abs(2*x)))
a = y[0:5:1]

print(y)
print(a)

plt.grid()
plt.xlabel("–ù–æ–º–µ—Ä —ç–ª–µ–º–µ–Ω—Ç–∞")
plt.ylabel("–ó–Ω–∞—á–µ–Ω–∏–µ")
plt.plot(range(1, 11), y)
plt.scatter(range(1, 6), a)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz

x = np.arange(11)
y = np.abs(np.cos(x*np.exp(np.cos(x) + np.log(x+1))))

plt.grid()
plt.plot(x, y)
plt.fill_between(x, y, alpha=0.4)

area = trapz(y)
print('–ü–ª–æ—â–∞–¥—å –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º: ', area)
```

4.3.5 –ó–∞–¥–∞–Ω–∏–µ: –°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –≤–∫–ª—é—á–∞—é—â–∏–π –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è
–¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö ‚Ä≤ + ‚Ä≤, ‚Ä≤ ‚àí ‚Ä≤, ‚Ä≤ ‚àô ‚Ä≤, ‚Ä≤ √∑ ‚Ä≤, –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ
—Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: ùëí^(ùë•+ùë¶)
, sin(ùë• + ùë¶) , cos(ùë• + ùë¶) , ùë•^ùë¶
.

```python
import math

print("–í—ã–±–µ—Ä–∏—Ç–µ –¥–µ–π—Å—Ç–≤–∏–µ:")
print("1. –°–ª–æ–∂–µ–Ω–∏–µ")
print("2. –í—ã—á–∏—Ç–∞–Ω–∏–µ")
print("3. –£–º–Ω–æ–∂–µ–Ω–∏–µ")
print("4. –î–µ–ª–µ–Ω–∏–µ")
print("5. e^(x+y)")
print("6. sin(x+y)")
print("7. cos(x+y)")
print("8. x^y")

while True:
    key = input()
    if (not key in ['1', '2', '3', '4', '5', '6', '7', '8']):
        print("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –≤–≤–æ–¥")
    else:
        break

key = int(key)
x = 0
y = 0

print("–í–≤–µ–¥–∏—Ç–µ x")
while True:
    x = input()
    try:
        valx = float(x)
        break
    except ValueError:
        print("–ü–æ–≤—Ç–æ—Ä–∏—Ç–µ –≤–≤–æ–¥")

print("–í–≤–µ–¥–∏—Ç–µ y")
while True:
    y = input()
    try:
        valy = float(y)
        break
    except ValueError:
        print("–ü–æ–≤—Ç–æ—Ä–∏—Ç–µ –≤–≤–æ–¥")

match key:
    case 1: print(valx + valy)
    case 2: print(valx - valy)
    case 3: print(valx * valy)
    case 4: print(valx / valy)
    case 5: print(math.exp(valx+valy))
    case 6: print(math.sin(valx+valy))
    case 7: print(math.cos(valx+valy))
    case 8: print(valx ** valy)
```
